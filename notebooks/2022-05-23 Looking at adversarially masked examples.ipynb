```python
table = artifact.get('adversarial_mask_table.table.json')

import pandas as pd
df = pd.DataFrame(table.data, columns=['name', 'document', 'masked_words', 'num_words_to_mask_next'])
df.head()
```

```python
from typing import List
import re

def fixed_redact_str(text: str, words_to_mask: List[str], mask_token='<mask>') -> str:
        for w in words_to_mask:
            text = re.sub(
                (r'\b{}\b').format(re.escape(w)),
                mask_token, text, count=0
            )
        return text

# fixed_redact_str(text=ex["document"], words_to_mask=self.adv_word_mask_map[idx])
```

```python
redacted_examples = df.apply(lambda ex: fixed_redact_str(text=ex['document'], words_to_mask=ex['masked_words']), axis=1)
```
